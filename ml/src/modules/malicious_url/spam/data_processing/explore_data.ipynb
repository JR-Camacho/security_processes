```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read the csv in a dataframe
df = pd.read_csv('../../../../../data/raw/urls/Spam.csv')
df.head()
```

```python
df.info()
```

```python
#Count the targets label
df["URL_Type_obf_Type"].value_counts()
```

```python
# Check for null values
is_null = df.isna().any()
is_null[is_null]
```

```python
# Check for infinite values
is_inf = df.isin([np.inf, -np.inf]).any()
is_inf[is_inf]
```

```python
# Graphical representation of two features
plt.figure(figsize=(12, 6))
plt.scatter(df["domainUrlRatio"][df['URL_Type_obf_Type'] == "spam"], df["domainlength"][df['URL_Type_obf_Type'] == "spam"], c="r", marker=".")
plt.scatter(df["domainUrlRatio"][df['URL_Type_obf_Type'] == "benign"], df["domainlength"][df['URL_Type_obf_Type'] == "benign"], c="g", marker="x")
plt.xlabel("domainUrlRatio", fontsize=13)
plt.ylabel("domainlength", fontsize=13)
plt.show()
```
